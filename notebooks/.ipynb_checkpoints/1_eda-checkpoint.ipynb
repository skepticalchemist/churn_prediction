```python
# importing libraries
import datetime as dt

import matplotlib.pyplot as plt
import missingno as msno
import numpy as np
import pandas as pd
import seaborn as sns
import sweetviz as sv
from pandas.plotting import scatter_matrix
```

### Importing the data

```python
# reading the dataset
churn = pd.read_csv("data.csv")
```

### Get to know the data

```python
# visualizing the first lines of the dataset
churn.head()
```

```python
# showing dimensions of the dataset
print("Dimensions of the dataset:")
print(f" Number of rows: {churn.shape[0]}")
print(f" Number of columns: {churn.shape[1]}")
print()
```

```python
# summarizing some info about the dataset
churn.info()
```

```python
# showing some decriptive statistics
churn.describe()
```

```python
# listing columns names
list(churn.columns)
```

```python
# visualize the dataset
plt.style.use("seaborn-deep")
scatter_matrix(
    churn, alpha=0.2, diagonal="hist", range_padding=0.4, marker="+", figsize=(18, 18)
)
plt.show()
plt.close()
```

```python
# create a pairplot using seaborn to visualize relationships in the dataset

sns.set(style="ticks")
# sns.set_context('poster')

sns.pairplot(churn, diag_kind="auto", kind="scatter", markers="+")
sns.color_palette("bright")
plt.suptitle("Complete dataset", y=1.05, size=50)
plt.show()
```

```python
# analyzing the dataset
sv_report = sv.analyze(churn)

# display the report
sv_report.show_html("report.html")
```

```python
# Correlation matrix completely ignores any non-numeric column.
fig, ax = plt.subplots(figsize=(6, 6))

matrix = np.triu(churn.corr("spearman"))
sns.set(font_scale=0.6)
sns.heatmap(
    churn.corr(method="spearman"),
    linewidth=0.1,
    annot=True,
    cmap="Blues",
    cbar_kws={"label": "correlation"},
    mask=matrix,
)
plt.xlabel("features", size=14)
plt.ylabel("features", size=14)
plt.title("Correlation Matrix", size=18)
plt.show()
```

### Convert date column to datatime format

```python
# convert object column to datetime format
churn["date"] = pd.to_datetime(churn["date"])
```

### Categorial and numerial features

```python
# separate numeric from categorical data
numeric_data = churn.select_dtypes(include=[np.number])
categorical_data = churn.select_dtypes(exclude=[np.number])

print(
    f" Number of numeric features: {numeric_data.shape[1]}\n Number of categorical features: {categorical_data.shape[1]}"
)
```

### Number of unique values

```python
# show unique values
for col in churn.columns:
    print(f"{col: >24}: {churn[col].nunique()}")
```

### Missing values

```python
# total number of null values, if they are present
print("Features with missing values:\n")
churn.isnull().sum()[churn.isnull().sum() > 0]
```

```python
# visualize missing value matrix for the complete dataset

msno.matrix(churn, labels=True, fontsize=12, figsize=(10, 2))
plt.title("Complete dataset features ", fontsize=24)
plt.show()
plt.close()
```
