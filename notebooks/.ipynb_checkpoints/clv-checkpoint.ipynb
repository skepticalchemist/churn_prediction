### Missing values

From the previous output we can set that the explainable variables `gender`, `vertical` have missing values. Lets take a closer look at this.

We have here at least to approaches, coding(using `isnull()`) or We can just visualize the missing values, in a more intuitive way, using the library `missingno`. The white lines in the `missingno` plot represent the missing values.

```python
# total number of missing values, if they are present
print("Features with missing values:\n")
churn.isnull().sum()[churn.isnull().sum() > 0]
```

The missing values correspond to:
* `gender` less than 1% of the dataset
* `vertical` around 3% of the dataset

```python
# visualize missing value matrix for the complete dataset
msno.matrix(churn, labels=True, fontsize=12, figsize=(10, 2))
plt.title("Visualizing missing values", fontsize=24)
plt.show()
plt.close()
```

### Visualizing the dataset with a pair-plot

We can plot a pairplot to visualize the relationships between features in the dataset.

```python
# visualize the dataset
plt.style.use("seaborn-deep")
sm = scatter_matrix(
    churn, alpha=0.3, diagonal="hist", range_padding=0.4, marker="+", figsize=(12, 12)
)

# change label rotation
[s.xaxis.label.set_rotation(45) for s in sm.reshape(-1)]
[s.yaxis.label.set_rotation(0) for s in sm.reshape(-1)]

# offset label when rotating to prevent overlap of figure
[s.get_yaxis().set_label_coords(-2.0, 0.5) for s in sm.reshape(-1)]
[s.get_xaxis().set_label_coords(0, -0.7) for s in sm.reshape(-1)]

# plt.tight_layout(h_pad=0.08, w_pad=0.08)
# plt.suptitle("scatter matrix plot", y=1.05, size=50)
plt.show()
plt.close()
```

### Correlation matrix

```python
# Correlation matrix completely ignores any non-numeric column.
fig, ax = plt.subplots(figsize=(6, 6))

matrix = np.triu(churn.corr("spearman"))
sns.set(font_scale=1.0)
sns.heatmap(
    churn.corr(method="spearman"),
    linewidth=0.1,
    annot=True,
    cmap="inferno",
    cbar_kws={"label": "correlation"},
    mask=matrix,
)
plt.xlabel("features", size=16)
plt.xticks(rotation=45, ha="right")
plt.ylabel("features", size=16)
plt.title("Correlation Matrix", size=18)
plt.show()
plt.close()
```

The strongest correlations occur between the features:
* `winning_cash_sum` and `turnover_cash_sum` (positive correlation)
* `NRG_sum` and `result_cash_sum` (negative correlation)

## Feature engineering

Let's first create a copy of the dateframe.

```python
churn_eng = churn.copy()
```

### Convert date column to datatime format

```python
# convert object column to datetime format
churn_eng["date"] = pd.to_datetime(churn_eng["date"])
```

### Create 4 new columns from `date`:
* `day_of_the_week`
* `day_of_the_month`
* `month`
* `week_of_year`

```python
churn.info()
```

```python
# create new features from date
churn_eng["day_of_week"] = churn_eng["date"].dt.weekday  # day of the week (0 = Monday)
churn_eng["day_of_month"] = churn_eng["date"].dt.day  # day of the month
churn_eng["month"] = churn_eng["date"].dt.month  # month
churn_eng["week_of_year"] = churn_eng["date"].dt.isocalendar().week  # week of the year
```

Modifying the `player_id`.

```python
# convert player_id to category to label enconder it
churn_eng["player_id"] = churn_eng["player_id"].astype("category")
churn_eng["player_id_cat"] = churn_eng["player_id"].cat.codes
churn_eng["player_id_cat"] = churn_eng["player_id_cat"] + 1
```
